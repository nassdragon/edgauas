# Read Data

```python
import pandas as  pd
df_pumpkin = pd.read_excel('Pumpkin_Seeds_Dataset.xlsx')
df_pumpkin
```

```python
df_pumpkin.info()
```

```python
df_pumpkin.describe() 
```

```python
pd.unique(df_pumpkin['Class'])
```

```python
df_pumpkin['Class'].value_counts()
```

## Label Encoder

```python
from sklearn.preprocessing import LabelEncoder as label_encoder
le = label_encoder()
df_pumpkin['Class'] = le.fit_transform(df_pumpkin['Class'])
df_pumpkin['Class'].value_counts()
```

## Split Atribute

```python
x = df_pumpkin.drop('Class', axis=1)
y = df_pumpkin['Class'] 
```

## Scalling

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(x) 
x = scaler.transform(x) 
```

## Split Train Test

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.15, random_state=33)
```

## Training Model

```python
from sklearn.linear_model import Perceptron

pcp_model = Perceptron()
pcp_model.fit(x_train, y_train)
pcp_model
```

## Predict Test

```python
y_pred = pcp_model.predict(x_test)
y_pred
```

## Acurate

```python
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report 

accuracy = accuracy_score(y_test, y_pred)  
report = classification_report(y_test, y_pred, target_names=le.classes_)
print("Akurasi Model:", accuracy)  
print("\nLaporan Klasifikasi:\n", report)
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

pcp_cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10, 7))
sns.heatmap(pcp_cm, annot=True, fmt='d', cmap='Blues', xticklabels=le.classes_, yticklabels=le.classes_)
plt.title('Confusion Matrix - Perceptron Model')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=le.classes_, y=pcp_cm.diagonal(), palette="viridis")
plt.title("Perceptron Prediction Class Distribution")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

## New Data

```python
new_data = pd.DataFrame({
    'Area': [67000, 72000],  
    'Perimeter': [950, 1000],  
    'Major_Axis_Length': [380, 400],  
    'Minor_Axis_Length': [220, 230], 
    'Convex_Area': [68000, 73000], 
    'Equiv_Diameter': [300, 310],  
    'Eccentricity': [0.8, 0.85],  
    'Solidity': [0.99, 0.98],  
    'Extent': [0.75, 0.73],  
    'Roundness': [0.85, 0.87],  
    'Aspect_Ration': [1.7, 1.8],  
    'Compactness': [0.75, 0.78]  
})
print(new_data)

new_predictions = pcp_model.predict(new_data)  
new_pumpkin = le.inverse_transform(new_predictions)  
for i, name in enumerate(new_pumpkin):
    print(f"Data baru {i+1}: Prediksi Class = {name}")
```
