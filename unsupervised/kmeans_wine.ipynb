### Data Training

```python
x = df_wine[['Alcohol', 'Total_Phenols']]
```

### Fitur Scaling

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
scale.fit(x)
x = scale.transform(x)
```

### Modeling,Evaluasi,dan Prediksi

```python
from sklearn.cluster import KMeans
kmean = KMeans(n_clusters=2)
kmean
```

```python
y_cluster = kmean.fit_predict(x)
y_cluster
```

```python
df_wine['cluster'] = y_cluster
df_wine
```

```python
kmean.cluster_centers_
```

### Visualisasi

```python
plt.scatter(x[:,0], x[:,1], c=y_cluster) 
plt.scatter(kmean.cluster_centers_[:,0], kmean.cluster_centers_[:,1], marker="*", s=250, c=[0,1], edgecolors="k") 

plt.xlabel("Alcohol")
plt.ylabel("Total Phenols")
plt.title("KMeans k=2")
plt.show() 
```

### Mencari nilai k yang terbaik

```python
kmean.inertia_
```

```python
inertias = []
k_range = range(1,11)
for k in k_range:
    kmean = KMeans(n_clusters=k)
    kmean.fit(x)
    inertias.append(kmean.inertia_)
    
plt.figure(figsize=(10,5))
plt.plot(k_range,inertias)
plt.plot(np.arange(1,11), inertias, marker="o")
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
plt.show() 
```
